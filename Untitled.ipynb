# **VINCENT MAKES A PITCH**

````python
import numpy as np
import matplotlib.pylab as plt 
from operator import add, sub, rshift, or_
````

````python
class Token:
    def __init__(self, name, vector):
        self.name = name
        self.vector = np.array(vector)
        
    def __add__(self, other):
        return self.__class__(name=f"({self.name} + {other.name})",
                              vector=self.vector + other.vector)

    def __sub__(self, other):
        return self.__class__(name=f"({self.name} - {other.name})",
                              vector=self.vector - other.vector)
    
    def __gt__(self, other):
        """
        return the scale when we map self unto other
        """
        return (self.vector.dot(other.vector))/(self.vector.dot(self.vector))

    def __rshift__(self, other):
        new_vec = (self.vector.dot(other.vector))/(self.vector.dot(self.vector))*self.vector
        return self.__class__(name=f"({self.name} >> {other.name})", vector=new_vec)
    
    def __or__(self, other):
        new_vec = self.vector - (other >> self).vector
        return self.__class__(name=f"({self.name} | {other.name})", vector=new_vec)
    
    def __repr__(self):
        return f"Token[{self.name}]"
    
    def plot(self, kind="scatter", x_axis=None, y_axis=None, color=None):
        """
        kind must be in [scatter, arrow, text]
        """
        if len(self.vector) == 2:
            handle_2d_plot(self, kind=kind, color=color)
        else:
            handle_nd_plot(self, kind=kind, x_axis=x_axis, y_axis=y_axis, color=color)
        return self

def handle_2d_plot(token, kind, color=None):
    if kind == "scatter":
        if color is None:
            color = "steelblue"
        plt.scatter([token.vector[0]], [token.vector[1]], c=color)
    if kind == "arrow":
        plt.quiver([0], [0], [token.vector[0]], [token.vector[1]], color=color,
             angles='xy', scale_units='xy', scale=1)
    if kind == "text":
        plt.text(token.vector[0] + 0.01, token.vector[1], token.name)
    plt.xlabel("x")
    plt.ylabel("y")

def handle_nd_plot(token, kind, x_axis=None, y_axis=None, color=None):
    x_val = token > x_axis
    y_val = token > y_axis
    intermediate = Token(name=token.name, vector=[x_val, y_val])
    handle_2d_plot(intermediate, kind=kind, color=color)
````

````python
class Embedding:
    """
    An embedding object that you can play with. 
    """
    def __init__(self, language, operations=list()):
        self.language = language
        self.operations = operations
    
    def operate(self, other, operation):
        return Embedding(language=self.language, 
                         operations=self.operations + [(other, operation)])
    
    def __add__(self, other):
        return self.operate(other, add)

    def __sub__(self, other):
        return self.operate(other, sub)

    def __or__(self, other):
        return self.operate(other, or_)

    def __rshift__(self, other):
        return self.operate(other, rshift)
    
    def __getitem__(self, thing):
        # if isinstance(thing, list):
        #    return [self[t] for t in thing]
        if isinstance(self.language, dict):
            return self.language[thing]
    
    def __repr__(self):
        result = "Embedding"
        translator = {add: '+', sub: '-', or_: '|', rshift: '>>'}
        for tok, op in self.operations:
            result = f"({result} {translator[op]} {tok.name})"
        return result
````

````python
class Embedding2D:
    """Two Dimensional Embedding Object"""
    def __init__(self, vector):
        self.vector = np.array(vector)
    
    def __add__(self, other):
        return self.__class__(self.vector + other.vector)

    def __and__(self, other):
        return self - other

    def __rshift__(self, other):
        new_vec = (self.vector.dot(other.vector))/(self.vector.dot(self.vector))*self.vector
        return self.__class__(new_vec)
    
    def __or__(self, other):
        new_vec = self.vector - (other >> self).vector
        return self.__class__(new_vec)

    def plot(self, color=None):
        plt.arrow(0, 0, self.vector[0], self.vector[1], head_width=0.02, head_length=0.1, fc=color, ec=color)
````

````python
plt.figure(figsize=(5, 5))
e1 = Embedding2D([0.4, 0.4])
e2 = Embedding2D([0.1, 0.2])
e3 = e1 >> e2
e4 = e2 | e1
e1.plot(color='blue')
e2.plot(color='orange')
e3.plot(color='red')
e4.plot(color='black')
plt.xlim(-.4, .6)
plt.ylim(-.4, .6)
````
